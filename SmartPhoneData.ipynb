```python
def graphs(col):
  sns.displot(kind='hist',kde=True,data=v3,x=col, label=col)
  sns.catplot(kind='box',data=v3,x=col)
```

```python
v3.select_dtypes(include=['float','int'])
```

```python
v3.select_dtypes(include=['float','int']).iloc[[3,4,6,913,14,16]].head()
```

```python
num_columns = v3.select_dtypes(include=['float','int']).iloc[[3,4,6,913,14,16]].columns
```

```python
for col in num_columns:
  graphs(col)
```
